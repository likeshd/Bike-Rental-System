### heatmap

```python
car_crash = sns.load_dataset('car_crashes')
corelations = car_crash.corr()
corelations
```

```python
sns.heatmap(corelations, annot=True, linewidths=0.5, cmap = 'cividis')
```

### Jointplot - both uni and bi

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.distplot(iris['sepal_length'])
```

```python
sns.jointplot(x = 'sepal_length', y = 'petal_length', data = iris)
```

```python
### decorators --
## adding new functionalities to existing python objects

def plus_one(number):
    result = number +1
    return result

plus_one(5)
```

```python

def math(number):
    def plus_one(number):
        result = number +1
        return result


    out = plus_one(number)
    result = out*2
    return result

math(5)
```
