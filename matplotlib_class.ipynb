#### Area chart

```python
a = range(0,5)
b = [[1,2,3,4,5],[2,3,4,5,6],[4,5,10,15,20]]

fig = plt.figure(figsize=(10,5))
ax1 = plt.subplot()
ax1.stackplot(a,b)
plt.title('Area chart demo')
plt.xlabel('a')
plt.ylabel('b')
```

### Vertical Bar chart

```python
data = {'apple':10,'banana':20,'apricot':5,'mango':15}
names = list(data.keys())
values = list(data.values())

fig = plt.figure(figsize = (10,5))

ax1 = plt.subplot(211)
ax2 = plt.subplot(212)
ax1.bar(names,values)
ax2.barh(names,values)
```

### Pie Chart

```python
names = ['Cat','Dog','Lion','Tiger','Elephant']
sizes = [10,15,5,20,30]
out = (0,0,0,0.2,0.1) # explode

plt.pie(x = sizes, labels = names, autopct='%1.1f%%', explode = out)
```

### Donut Chart

```python
main_group = ['A','B','C']
main_size = [10,20,30]

subgroups = ['A.1','A.2','B.1','B.2','B.3','C.1','C.2','C.3','C.4','C.5']
subsize = [6,4,10,6,4,10,8,6,4,2]

a,b,c = [plt.cm.Blues, plt.cm.Reds, plt.cm.Greens]

fig,ax = plt.subplots()
ax.axis('equal')

pie1, _ = ax.pie(main_size,labels = main_group, radius = 1.5, labeldistance=0.9, colors = [a(0.8),b(0.8),c(0.8)])
pie2, _  = ax.pie(subsize,labels = subgroups, radius = 1.2, labeldistance=0.7, colors = [a(0.6),a(0.5),b(0.6),b(0.5),b(0.4),c(0.6),c(0.5),c(0.4),c(0.3),c(0.2),c(0.1)])

plt.setp(pie1, width = 0.3, edgecolor = 'white')
plt.setp(pie2, width = 0.4, edgecolor = 'white')
```
